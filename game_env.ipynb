```python
# Import necessary libraries for Boxing environment
import gymnasium as gym
import ale_py
import numpy as np
import matplotlib.pyplot as plt

# Register ALE environments with Gymnasium
gym.register_envs(ale_py)

print("Available ALE environments registered successfully!")
print(f"ALE-py version: {ale_py.__version__}")
print(f"Gymnasium version: {gym.__version__}")
```

```python
# Create the Boxing environment
env_name = "ALE/Boxing-v5"
env = gym.make(env_name, render_mode="rgb_array")
print(f"Environment: {env_name}")
print(f"Action space: {env.action_space}")
print(f"Observation space: {env.observation_space}")
print(f"Action meanings: {env.get_action_meanings()}")

# Reset the environment to get initial observation
observation, info = env.reset()
print(f"Initial observation shape: {observation.shape}")
print(f"Initial info: {info}")

# Display the initial frame
plt.figure(figsize=(8, 6))
plt.imshow(observation)
plt.title("Boxing Environment - Initial Frame")
plt.axis('off')
plt.show()
```
